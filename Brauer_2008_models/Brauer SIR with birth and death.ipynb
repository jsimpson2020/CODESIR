# Brauer SIR model with birth and death

Assumptions:

$S(t)$ denotes the number of individuals who are susceptible to the disease, that is, who are not (yet) infected at time $t$.

$I(t)$ denotes the number of infected individuals, assumed infectious and able to spread the disease by contact with susceptibles.

$R(t)$ denotes the number of individuals who have been infected and then removed from the possibility of being infected again or of spreading infection. Removal is carried out either through isolation from the rest of the population or through immunization against infection or through recovery from the disease with full immunity against reinfection or throuhg death caused by the disease.

The number of members in a compartment is a differentiable function of time. This may be a reasonable approximation if there are many members in a compartment, but it is certainly suspect otherwise.

The epidemic process is deterministic, that is, that the behaviour of a population is determined completely by its history and by the rules which describe the model.

To incorporate disease death we must include an equation for total population size in the model. This forces us to make a distinction between members of the population who die of the disease and members of the population who recover with immunity against reinfection.

The differential equations for this situation are:
        $$S'=\Lambda(N)-\beta(N)SI-\mu S$$
        $$I'=\beta(N)SI-\mu I-\alpha I$$
        $$R'=f\alpha I-\mu R$$
        $$N'=\Lambda(N)-(1-f)\alpha I-\mu N$$
        
These include the following assumptions:

(1) An average member of the population makes contact sufficient to transmit infection with $\beta(N)N$ others per unit time, where $N$ represents total population size. If $\beta$ is a constant, then we have mass action incidence. We will let $\beta(N)=\frac{C(N)}{N}$, where $C(N)$ is the number of contacts an average member of the population makes in unit time. We will be using $C(N)=\lambda$, which is referred to as standard incidence.

(2) Infectives leave the infective class at rate $\alpha I$ per unit time.

(3) A fraction $f$ of the $\alpha I$ members leaving the infective class at time $t$ recover and the remaining fraction $(1-f)$ die of disease.
        
(4) The birth rate is a function $\Lambda(N)$ of total population size $N$. To keep the natural population constant let $\Lambda(N)=\mu N$. We will let $\Lambda(N)=\nu N$, where $\nu$ is the natural birth rate in a population per person per unit time.

(5) Death rate is the natural death rate $\mu$ in each class.

The differential equations will more specifically look like:
        $$S'=\nu N-\frac{\lambda}{N}SI-\mu S$$
        $$I'=\frac{\lambda}{N}SI-\mu I-\alpha I$$
        $$R'=f\alpha I-\mu R$$
        $$N'=\nu N-(1-f)\alpha I-\mu N$$

### Importing Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy as sp
import pandas as pd
```

### Initializing Parameters

```python
dt = 0.01
tottime = 3650
lmbda = 0.25
alpha = 0.07
mu = 0.001
nu = 1.01*mu
f = 0.95
N0 = 10000
```

### Initializing Arrays

```python
niter = int(math.ceil(tottime/dt))
t = np.arange(0, tottime, dt)   
S = np.zeros(niter)
I = np.zeros(niter)
R = np.zeros(niter)
N = np.zeros(niter)
```

### Initializing Populations

```python
pi = 0.01
pr = 0
ps = 1-pi-pr

S[0] = ps*N0
I[0] = pi*N0
R[0] = pr*N0
N[0] = N0
```

### Euler's Method

```python
for j in range(niter-1):
    dSdt=nu*N[j]-lmbda/N[j]*S[j]*I[j]-mu*S[j]
    dIdt=lmbda/N[j]*S[j]*I[j]-mu*I[j]-alpha*I[j]
    dNdt=nu*N[j]-(1-f)*alpha*I[j]-mu*N[j]
    S[j+1] = S[j] + dt*dSdt 
    I[j+1] = I[j] + dt*dIdt
    N[j+1] = N[j] + dt*dNdt
   
       
R = N-S-I
```

### Plotting

```python
plt.plot(t, S, 'k')
plt.plot(t, I, 'm')
plt.plot(t, R, 'b')
plt.plot(t, N, 'y')
plt.gca().legend(('susceptible','infected','recovered','total'),loc='right')
plt.title('Brauer 2008 SIR model with birth and death')
plt.xlabel('days since 1 percent of population infected')
plt.ylabel('population')
plt.show()
```

### Interactive model with intuitive variables

```python
@interact(total_time=(100,50000,5), avg_contacts_per_day=(0,5,0.01), length_of_infection=(0,100,1),
          deaths_per_100000_per_day=(0,100000,0.1), births_per_100000_per_day=(0,1000,0.1),
          fraction_recovered=(0,1,0.01), initial_population=(0,1000000,1000),
          initial_percent_recovered=(0,100,0.01), initial_percent_infected=(0,100,0.01),
         xmin=(0,10000,5), xmax=(0,50000,5), ymin=(0,100000,50), ymax=(0,2000000,50))
def SIR_model(total_time=4000, avg_contacts_per_day=0.50, length_of_infection=31,
              deaths_per_100000_per_day=2.4, births_per_100000_per_day=3.3,
              fraction_recovered=1.0, initial_population=100000,
              initial_percent_recovered=0.0, initial_percent_infected=1.0,
             xmin=0, xmax=4000, ymin=0, ymax=100000):
    dt = 0.01
    
    lmbda = avg_contacts_per_day
    
    alpha = 1/length_of_infection
    
    mu = deaths_per_100000_per_day/100000
    
    nu = births_per_100000_per_day/100000
    
    f = fraction_recovered
    
    niter = int(math.ceil(total_time/dt))
    t = np.arange(0, total_time, dt)   
    S = np.zeros(niter)
    I = np.zeros(niter)
    R = np.zeros(niter)
    N = np.zeros(niter)
    
    pi = initial_percent_infected/100
    pr = initial_percent_recovered/100
    ps = 1-pi-pr

    S[0] = ps*initial_population
    I[0] = pi*initial_population
    R[0] = pr*initial_population
    N[0] = initial_population
    
    for j in range(niter-1):
        dSdt=nu*N[j]-lmbda/N[j]*S[j]*I[j]-mu*S[j]
        dIdt=lmbda/N[j]*S[j]*I[j]-mu*I[j]-alpha*I[j]
        dNdt=nu*N[j]-(1-f)*alpha*I[j]-mu*N[j]
        S[j+1] = S[j] + dt*dSdt 
        I[j+1] = I[j] + dt*dIdt
        N[j+1] = N[j] + dt*dNdt
        if I[j] < 1:
            I[j+1]=0
        if N[j] < 1:
            S[j+1]=0
            I[j+1]=0
            R[j+1]=0
    R = N-S-I
    
    R0 = lmbda/(mu + alpha)
    
    R0approx = round(R0, 3)
    
    plt.plot(t, S, 'k')
    plt.plot(t, I, 'm')
    plt.plot(t, R, 'b')
    plt.plot(t, N, 'y')
    plt.gca().legend(('susceptible','infected','recovered','total'),loc='right')
    plt.title('Brauer 2008 SIR model with birth and death, R0 = ' + str(R0approx))
    plt.xlabel('days since ' + str(initial_percent_infected) +
               ' percent of population infected')
    plt.ylabel('population')
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    plt.show()
    
    print('Total time = ' + str(total_time) + ' days',
          'Average contacts with an infective sufficient to transmit disease per day = '
          + str(lmbda) + ' people',
          'Length of infection = ' + str(length_of_infection) + ' days',
          'Deaths per 100000 per day = ' + str(deaths_per_100000_per_day) + ' deaths',
          'Births per 100000 per day = ' + str(births_per_100000_per_day) + ' births',
          'Fraction of population who recover = ' + str(fraction_recovered),
          'Initial population = ' + str(initial_population) + ' people',
          'Initial percent recovered = ' +str(initial_percent_recovered) + '%',
          'Initial percent infected = ' + str(initial_percent_infected) + '%',
         sep = "\n")
    
    return plt.show()
```

### Adding in re-susceptibility (SIRS)

Further assumptions:

Individuals in the recovered population do not have immunity forever. We will allow them to lose their immunity at a rate $r$. Hence, the differential equations become:

$$S'=\nu N-\frac{\lambda}{N}SI-\mu S+rR$$
        
$$I'=\frac{\lambda}{N}SI-\mu I-\alpha I$$
        
$$R'=f\alpha I-\mu R-rR$$
        
$$N'=\nu N-(1-f)\alpha I-\mu N$$

We assume that each individual loses immunity at the same average rate. Note that $r=\frac{1}{\text{length of immunity in days}}$.

```python
@interact(total_time=(100,10000,5), avg_contacts_per_day=(0,5,0.01), length_of_infection=(0,50,1),
          deaths_per_100000_per_day=(0,100000,0.1), births_per_100000_per_day=(0,1000,0.1),
          fraction_recovered=(0,1,0.01), initial_population=(0,500000,1000),
          initial_percent_recovered=(0,100,0.01), initial_percent_infected=(0,100,0.01),
          length_of_immunity=(1,5000,1), xmin=(0,10000,5), xmax=(0,10000,5),
          ymin=(0,100000,50), ymax=(0,600000,50))
def SIR_model(total_time=4000, avg_contacts_per_day=0.25, length_of_infection=20,
              deaths_per_100000_per_day=2.4, births_per_100000_per_day=3.3,
              fraction_recovered=1, initial_population=100000,
              initial_percent_recovered=0.0, initial_percent_infected=1.0,
              length_of_immunity=100, xmin=0, xmax=4000, ymin=0, ymax=110000):
    dt = 0.01
    
    lmbda = avg_contacts_per_day
    
    alpha = 1/length_of_infection
    
    mu = deaths_per_100000_per_day/100000
    
    nu = births_per_100000_per_day/100000
    
    f = fraction_recovered
    
    r = 1/length_of_immunity
    
    niter = int(math.ceil(total_time/dt))
    t = np.arange(0, total_time, dt)   
    S = np.zeros(niter)
    I = np.zeros(niter)
    R = np.zeros(niter)
    N = np.zeros(niter)
    
    pi = initial_percent_infected/100
    pr = initial_percent_recovered/100
    ps = 1-pi-pr

    S[0] = ps*initial_population
    I[0] = pi*initial_population
    R[0] = pr*initial_population
    N[0] = initial_population
    
    for j in range(niter-1):
        dSdt=nu*N[j]-lmbda/N[j]*S[j]*I[j]-mu*S[j]+r*R[j]
        dIdt=lmbda/N[j]*S[j]*I[j]-mu*I[j]-alpha*I[j]
        dRdt=f*alpha*I[j]-mu*R[j]-r*R[j]
        dNdt=nu*N[j]-(1-f)*alpha*I[j]-mu*N[j]
        S[j+1] = S[j] + dt*dSdt 
        I[j+1] = I[j] + dt*dIdt
        R[j+1] = R[j] + dt*dRdt
        N[j+1] = N[j] + dt*dNdt
        if I[j] < 1:
            I[j+1]=0
        if N[j] < 1:
            S[j+1]=0
            I[j+1]=0
            R[j+1]=0
    
    R0 = lmbda/(mu + alpha)
    
    R0approx = round(R0, 3)
    
    plt.plot(t, S, 'k')
    plt.plot(t, I, 'm')
    plt.plot(t, R, 'b')
    plt.plot(t, N, 'y')
    plt.gca().legend(('susceptible','infected','recovered','total'),loc='right')
    plt.title('SIR model with birth, death and temporary immunity \n R0 = ' + str(R0approx))
    plt.xlabel('days since ' + str(initial_percent_infected) +
               ' percent of population infected')
    plt.ylabel('population')
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    plt.show()
    
    print('Total time = ' + str(total_time) + ' days',
          'Average contacts with an infective sufficient to transmit disease per day = '
          + str(lmbda) + ' people',
          'Length of infection = ' + str(length_of_infection) + ' days',
          'Deaths per 100000 per day = ' + str(deaths_per_100000_per_day) + ' deaths',
          'Births per 100000 per day = ' + str(births_per_100000_per_day) + ' births',
          'Fraction of population who recover with temporary immunity = '
          + str(fraction_recovered),
          'Length of immunity = ' + str(length_of_immunity) + ' days',
          'Initial population = ' + str(initial_population) + ' people',
          'Initial percent recovered = ' +str(initial_percent_recovered) + '%',
          'Initial percent infected = ' + str(initial_percent_infected) + '%',
         sep = "\n")
    
    return plt.show()
```
